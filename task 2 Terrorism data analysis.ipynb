#### Renaming columns 

```python
dataframe = pd.DataFrame(terrorism_data, terrorism_data.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
dataframe.head()
```

#### Feature creation

#### Here will add new column as combination of no of people killed and injured in terrorist attack.

```python
dataframe['Total_Deaths'] = dataframe['Killed'] + dataframe['Wounded']
```

```python
dataframe =  dataframe[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude','Location','AttackType','Killed','Wounded','Total_Deaths','Target','Summary','Group','Target_type','Weapon_type','Motive', 'Success']]
```

####  Visualization between year and no of terrorist attack 

```python

data_Y_and_no_of_attack = []

for i in dataframe['Year'].unique():
    data_Y_and_no_of_attack.append([i , len(dataframe[dataframe['Year'] ==i])])
    
pd.DataFrame(data_Y_and_no_of_attack,columns = ['Year','Total Attacks']).plot(x = "Year", y = "Total Attacks", title='Number of Terrorists Activities Per Year',kind = 'bar',color='pink',figsize = (17,8))
m.style.use('dark_background')
m.legend()
```

### Conclusion : 

* Generally, as the Time proceeds the Number of Terrorists Attacks also Increased.
* The peak in terms of Number of Terrorists Attacks was observed in 2014.
* A Decline in terms of Number of Terrorists Attacks has been observed since 2015 which shows the progress towards the mission of reducing such terrorists attacks.

```python
casualities_year=dataframe.groupby('Year')['Total_Deaths'].sum()[0:]
```

```python
casualities_year.plot.bar(xlabel='Year',ylabel='Casualities (Killed + Injured)',title='Total Casualties Per Year',color='orange',figsize=(17,8))
m.style.use('dark_background')
m.legend()
```

### Conclusion:

* Generally, as the Time proceeds the Number of People Harmed (Killed or Injured) also Increased Significantly.

* The peak in terms of Number of People Harmed was observed in 2015 with 70K+ casualties.

* Though there has been a decline in the Number of People Harmed after 2016 but these numbers are still very large and need to be reduced as soon as possible.

### visualization between Number of terrorists activities and Countries

```python
country = dataframe['Country'].value_counts().index.tolist()
print("Number of Countries affected by Terrorist Attacks :", len(country))
```

```python
data_coun_acti = []

for i in dataframe['Country'].value_counts()[0:30].index:
    data_coun_acti.append([i , len(dataframe[dataframe['Country'] ==i])])
    
pd.DataFrame(data_coun_acti,columns = ['Country','Total Attacks']).sort_values(by='Total Attacks', ascending=False).plot(x = "Country", y = "Total Attacks", title='Countries affected by Terrorist Attacks',kind = 'bar',color='red',figsize = (17,8))
m.style.use('dark_background')
m.legend()
```

### Conclusion:

* The above image shows the Top 30 countries in terms of most number of terrorist attacks.

* The top country which is most affected by terrorist attacks is Iraq.

* The top 5 countries most affected by terrorist attacks are Iraq, Pakistan, Afganistan, India, and Colombia

### Top 30 Countries in terms of Causalities

```python
casua_country = dataframe.groupby('Country')['Total_Deaths'].sum().to_frame().reset_index().sort_values('Total_Deaths', ascending=False)
casua_country=casua_country.head(30)
casua_country
```

```python
casua_country.plot.bar(xlabel='Country',ylabel='Total_Causalities',title='Number of Total Casualities in Each Country',color='yellow',figsize=(17,8))
m.style.use('dark_background')
m.legend()
```

### Conclusion : 

* The above image shows the Most Number of Total Casualities in Each Country.

* The most affected country by Total Casualities is Iraq with 200K+ Casualities.

* The top 5 countries most affected by terrorist attacks are Iraq, Afganistan, Pakistan, India, Sri Lanka.

#### Most used types of attack by terrorist 

```python
attack_types_by_t = dataframe['AttackType'].value_counts()
print("Number of types of attacks executed by Terrorists :", len(attack_types_by_t))
```

```python
data_of_types_of_attack = []

for i in dataframe['AttackType'].value_counts().index:
    data_of_types_of_attack.append([i , len(dataframe[dataframe['AttackType'] ==i])])
    
plot_data=pd.DataFrame(data_of_types_of_attack,columns = ['AttackType','Total Attacks']).sort_values(by='Total Attacks', ascending=False)
plot_data.plot(x = "AttackType", y = "Total Attacks", title='Most Used Attack Types by Terrorist',kind = 'bar',color='purple',figsize = (17,8))
m.style.use('dark_background')
m.legend()
```

### Conclusion :-

•The above image shows the Most Used Attack Types by Terrorists.

•The Most Used Attack Type by Terrorists is Bombing/Explosion.

•The top 5 Attack Types used by Terrorists are Bombing/Explosion, Armed Assault, Assassination, Hostage Taking (Kidnapping), Facility/Infrastructure Attack

```python
data=[]
for i in dataframe['Target_type'].value_counts().index:
    data.append([i , len(dataframe[dataframe['Target_type'] ==i])])
    
plot_data=pd.DataFrame(data,columns = ['Target_type','Total Attacks']).sort_values(by='Total Attacks', ascending=False)
plot_data.plot(x = "Target_type", y = "Total Attacks", title='Most frequent target types of terrorists',kind = 'bar',color='yellow',figsize = (17,8))
m.style.use('dark_background')
m.legend()
```
